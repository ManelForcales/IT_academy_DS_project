```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df['types'] = df.types.map({'Male':1,'Female':1,'Ape':0,'Zombie':0,'Alien':0})
df.head()
```

```python
def convert_str2list(text):
    #print(text)
    try:
        items = text.split(',')        
        return items
    except:
        return 'None'
    
#df1.traits.apply(convert_str2list)
convert_str2list('Blonde Bob,Earring,Green Eye Shadow')
```

```python
df.traits = df['traits'].apply(convert_str2list)
```

```python
df.head()
```

```python
df_exploded =df.explode('traits')
df_exploded.reset_index(inplace=True)
df_exploded['_helper'] = 1
df_exploded
```

```python
df_wide = df_exploded.pivot(index='index', columns='traits',values='_helper')
df_wide.fillna(0,inplace=True)
df_wide
```

```python
df1 = pd.concat([df,df_wide],axis=1)
df1.drop('traits',axis=1,inplace=True)
df1
```

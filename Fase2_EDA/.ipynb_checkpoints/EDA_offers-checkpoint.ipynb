```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../offers.csv')
df.punk_id = df.punk_id.astype(int)
df.drop('Unnamed: 0',axis=1,inplace=True)
df['Txn'] = df['Txn'].astype('datetime64')
#df.set_index('punk_id',inplace=True)
df.head()
```

```python
df.shape
```

```python
punks = pd.read_csv('../Fase1_ETL/to_database/raw_punks.csv')
punks.drop('Unnamed: 0',axis=1,inplace=True)
#punks.set_index('id',inplace=True)
```

```python
punks.head()
```

```python
punks.shape
```

```python
aggregated_df = df.merge(punks, how='left',left_on='punk_id',right_on='id',left_index=False, right_index=False, sort=False)
aggregated_df.head()
```

```python
aggregated_df.Type.unique()
```

```python
# get punks who have been sold at least once
```

```python
aggregated_df['Sold'] = np.where((aggregated_df['Type']=="Sold"), 1, 0)
```

```python
filter_dataset = aggregated_df[aggregated_df.Sold == 1]['punk_id']
```

```python
selected_punks = filter_dataset.unique().tolist()
```

```python
# add column where to keep those punks of not
```

```python
aggregated_df['keep']= aggregated_df.punk_id.apply(lambda x: x in selected_punks)
```

```python
aggregated_df.head()
```

```python
# keep those punks only
aggregated_df = aggregated_df[aggregated_df.keep == True]
```

```python
aggregated_df.Type.unique()
```

```python
offers = aggregated_df[aggregated_df['Type']=='Offered']
offers.head()
```

```python
df_offer = offers[['From','To','Txn','punk_id','Type','amount_ethereum','types','skin','rarity','traits']].reset_index(drop=True)
```

```python
df_offer.head()
```

```python
df_offer.Type.unique()
```

```python
plt.subplots(figsize=(16,9))
df_offer.groupby('punk_id')['Type'].count().plot()
plt.title ('number of offers')
plt.show()
```

```python
df_count_offers = df_offer.groupby('punk_id')['Type'].count().to_frame(name='total_offers')
df_mean_price = df_offer.groupby('punk_id')['amount_ethereum'].mean().to_frame(name='avg_offer_price')
```

```python
df_counts = pd.concat([df_count_offers,df_mean_price,punks],axis=1)
df_counts = df_counts[['total_offers','avg_offer_price','types','rarity','skin']]
```

```python
#df_counts.fillna(0,inplace=True)
# remove item with NA as they do not exist in this subset

df_counts.dropna(inplace=True)
df_counts.head(10)
```

```python
df_raros = df_counts[df_counts.rarity<101].sort_values(by='rarity')
df_otros = df_counts[df_counts.rarity>101].sort_values(by='rarity')
```

```python
df_raros['tipo'] = df_raros.types.map({'Male':'humano','Female':'humano','Ape':'rest','Zombie':'rest','Alien':'rest'})
df_raros.head(100)
```

```python
# aquellos que tienen 1 o mas pujas
df_raros = df_raros[df_raros['total_offers']>0]
df_raros.total_offers = df_raros.total_offers.astype(int)
df_raros.head(5)
```

```python
fig,ax = plt.subplots(figsize=(12,8))
sns.histplot(data=df_raros['total_offers'],bins=50,stat='probability',ax=ax,discrete=True)
plt.title ('TOP 100 RAREST PUNKS; nº TOTAL OFFERS HISTOGRAM')
plt.show()
df_raros
```

```python
def outliers(df, feature):
    Q1= df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR
    return upper_limit, lower_limit
```

```python
mediana = df_raros['avg_offer_price'].median()
std = df_raros['avg_offer_price'].std()
perc_75 = df_raros['avg_offer_price'].quantile(.95)
print(mediana,std,perc_75)
upper, lower = outliers(df_raros, "avg_offer_price")
print(upper,lower)
```

```python
df_otros = df_counts.sort_values(by='rarity')[100:]
```

```python
mediana = df_otros['avg_offer_price'].median()
std = df_otros['avg_offer_price'].std()
perc_75 = df_otros['avg_offer_price'].quantile(0.75)
print(mediana,std,perc_75)
upper, lower = outliers(df_otros, "avg_offer_price")
print(upper,lower)
```

```python
fig,ax = plt.subplots(figsize=(10,6))
#display(df_otros.head())
#display(df_otros.types.unique())
df_otros['tipo'] = df_otros.types.map({'Male':'humano','Female':'humano','Ape':'rest','Zombie':'rest','Alien':'rest'})
sns.boxplot(x="tipo", y="avg_offer_price",data=df_otros[~(df_otros['avg_offer_price']>perc_75)],ax=ax)
plt.title('rest of punks: outliers removed')
plt.ylim([0,500])
plt.show()
```

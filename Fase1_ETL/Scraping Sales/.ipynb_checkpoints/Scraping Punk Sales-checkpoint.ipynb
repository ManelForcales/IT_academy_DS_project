```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import re
from IPython.display import Image
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

import geckodriver_autoinstaller
```

### Iterating between PUNK_ID web segments and tracking evey purchase transaction. First 1000 punk_id

```python
total_punks = (5000,5001)
amount_eth = []
amount_dollars = []
transaction_date= []
punk_id = []


for j in range(5000,6001):
    try:
        browser = webdriver.Chrome(executable_path=r"C:\Users\tomi_\.wdm\drivers\chromedriver\win32\91.0.4472.101\chromedriver.exe")
        url = 'https://www.larvalabs.com/cryptopunks/details/'+str(j)
        browser.get(url)
        browser.implicitly_wait(15)

        table = browser.find_element_by_tag_name('table')
        rows = table.find_elements_by_class_name('punk-history-row-sold')

        for row in rows:
            punk_id.append(j)
            items = row.text.split(' ')
            amount_eth.append(items[3])
            amount_dollars.append(items[4])
            transaction_date.append(' '.join(items[5:]))

        browser.close()



    except Exception as err:
        print(f'exepcion {err}')

#df = pd.DataFrame({'id':punk_id,'amount':amount,'transaction':transaction_date})
```

```python
len(amount_dollars)
```

### Creating DF

```python
df1 = pd.DataFrame({'id':punk_id,'amount_ethereum':amount_eth, 'amount_dollars': amount_dollars, 'transaction_date':transaction_date})
```

```python
df1
```

```python
df1.to_csv('df5000-6001.csv')
```

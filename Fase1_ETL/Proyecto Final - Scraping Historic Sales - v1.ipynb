### Defining workspace and URL to scrape

```python
driver = webdriver.Chrome(executable_path=r"C:\Users\tomi_\.wdm\drivers\chromedriver\win32\91.0.4472.101\chromedriver.exe")
driver.get("https://defypunk.com/punks/sales")
```

### Looping through rows and storing rows as list of rows (first 20 elements)

```python
# Iteración apendizando cada "fila" para luego ser divididas en columnas

rows = []
transaction_date = []

for i in range(20):
    values = []
    driver.implicitly_wait(20)
    x_path = ('//*[@id="root"]/div/main/div/div[2]/div/div/div[1]/div[2]/div[3]/div[2]/div/div/div'+str('[')+str(i)+str(']'))
    row = driver.find_elements_by_xpath(x_path)
    for y in row:
        values.append(y.text)
    rows.append(values)

rows
```

### Splitting rows into multiple columns. 

```python
new = []
for l in rows:
    new.append([x.split('\n') for x in l])

new
```

### Flattening of lists and conversion to DF

```python
flattened = []
for sublist in new:
    for val in sublist:
        flattened.append(val)
        
flattened
```

### Creating DF based on scraped rows

```python
sales_df = pd.DataFrame(flattened, columns=['punk_id', 'rarity', 'type', 'skin', 'traits', 'transaction_date', 'sale_price'])
sales_df
```

```python
sales_df.to_csv('sales_df.csv')
```
